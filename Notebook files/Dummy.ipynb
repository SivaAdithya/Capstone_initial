```python
activity = ['Lying', 'Self Pace walk', 'Running', 'Sitting']
```

```python
##age
age = []
gender = []
height = []
weight = []
heart_rate = []
rest_heart_rate = []
activity_trimmed = []
for i in range(len(data)):
    age.append(random.randint(50,90))
    gender.append(random.choice([1,0]))
    h = random.uniform(150,170)
    height.append(h)
    weight.append(random.uniform(27,30)*(h/100)*(h/100))
    ac = random.choice(activity)
    heart_rate.append(random.uniform(65,140))
    rest_heart_rate.append(random.uniform(60,90))
    activity_trimmed.append(ac)
```

```python
out = [1]*len(data)
```

```python
for i in data:
    print(i,max(data[i]))
```

```python
arr = {'Age':age,'Gender':gender,'Height':height,'Weight':weight,'Heart_rate':heart_rate,'Rest_heart_rate':rest_heart_rate,'Activity':activity_trimmed,'Output':out}
data_new = pd.DataFrame(arr)
```

```python
out = [0]*len(data2)
```

```python
arr = {'Age':age,'Gender':gender,'Height':height,'Weight':weight,'Heart_rate':heart_rate,'Rest_heart_rate':rest_heart_rate,'Activity':activity_trimmed,'Output':out}
data_new = pd.DataFrame(arr)
```
